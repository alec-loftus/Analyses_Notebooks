```python
#RANDOMIZES, GETS ENERGIES, SAVES ORIGINAL AS FRAME 1
import pyrosetta
from pyrosetta import pose_from_pdb, get_score_function, MonteCarlo
from pyrosetta.rosetta.core.select.residue_selector import (
    ResidueIndexSelector, NeighborhoodResidueSelector, NotResidueSelector, ResiduePropertySelector
)
from pyrosetta.rosetta.core.chemical import ResidueProperty
from pyrosetta.rosetta.core.pack.task import TaskFactory
from pyrosetta.rosetta.core.pack.task.operation import (
    InitializeFromCommandline, IncludeCurrent, NoRepackDisulfides,
    PreventRepackingRLT, OperateOnResidueSubset,
    RestrictToRepackingRLT, RestrictAbsentCanonicalAASRLT
)
from pyrosetta.rosetta.protocols.minimization_packing import PackRotamersMover
from pyrosetta.rosetta.core.pose import Pose
from pyrosetta.rosetta.core.id import TorsionID, BB
from pyrosetta.rosetta.std import stringstream

import random
import csv

# === Init PyRosetta ===
pyrosetta.init("-ex1 -ex2 -use_input_sc -flip_HNQ -mute all")

def pose_to_string(pose: Pose) -> str:
    ss = stringstream()
    pose.dump_pdb(ss)
    return ss.str()

def randomize_sidechains_in_subset(pose, residue_indices):
    for resi in residue_indices:
        if not pose.residue(resi).is_protein():
            continue
        nchi = pose.residue(resi).nchi()
        for chi in range(1, nchi + 1):
            pose.set_chi(chi, resi, random.uniform(0, 360))

def perturb_backbone(pose, residue_indices, max_angle=2.0):
    """Small backbone chi randomization for neighbors."""
    for resi in residue_indices:
        if not pose.residue(resi).is_protein():
            continue
        for torsion in [1, 2]:  # phi, psi
            original = pose.torsion(TorsionID(resi, BB, torsion))
            perturb = original + random.uniform(-max_angle, max_angle)
            pose.set_torsion(TorsionID(resi, BB, torsion), perturb)

def pack_and_save_multiple_rotamers(
    pose, posi, amino, scorefxn,
    n_structs=30,
    out_pdb_path="multi_rotamers.pdb",
    out_csv_path="energies.csv",
    mc_temp=2.0,
    apply_backbone_perturb=True
):
    mut_posi = ResidueIndexSelector()
    mut_posi.set_index(posi)

    nbr_selector = NeighborhoodResidueSelector()
    nbr_selector.set_focus_selector(mut_posi)
    nbr_selector.set_include_focus_in_subset(True)

    not_design = NotResidueSelector(mut_posi)

    tf = TaskFactory()
    tf.push_back(InitializeFromCommandline())
    tf.push_back(IncludeCurrent())
    tf.push_back(NoRepackDisulfides())

    prevent_repacking_rlt = PreventRepackingRLT()
    prevent_subset_repacking = OperateOnResidueSubset(prevent_repacking_rlt, nbr_selector, True)
    tf.push_back(prevent_subset_repacking)

    tf.push_back(OperateOnResidueSubset(RestrictToRepackingRLT(), not_design))

    aa_to_design = RestrictAbsentCanonicalAASRLT()
    aa_to_design.aas_to_keep(amino)
    tf.push_back(OperateOnResidueSubset(aa_to_design, mut_posi))

    mc = MonteCarlo(pose, scorefxn, temperature=mc_temp)

    pose_list = []
    energy_list = []

    # === Include Original Pose First ===
    original_pose = pose.clone()
    original_energy = scorefxn(original_pose)

    pose_list.append(original_pose.clone())
    energy_list.append({
        "Index": 1,
        "Energy_Before_MC": original_energy,
        "Energy_After_MC": original_energy
    })

    # === Generate Variant Conformations ===
    for i in range(1, n_structs + 1):  # i starts at 1, Index = i + 1
        tmp_pose = pose.clone()

        # Neighboring residues
        neighbor_subset = nbr_selector.apply(tmp_pose)
        neighbor_indices = [idx + 1 for idx, selected in enumerate(neighbor_subset) if selected]

        randomize_sidechains_in_subset(tmp_pose, neighbor_indices)
        if apply_backbone_perturb:
            perturb_backbone(tmp_pose, neighbor_indices, max_angle=2.0)

        packer = PackRotamersMover()
        packer.task_factory(tf)
        packer.score_function(scorefxn)
        packer.apply(tmp_pose)

        energy_before_mc = scorefxn(tmp_pose)

        mc.reset(tmp_pose)
        mc.boltzmann(tmp_pose)
        energy_after_mc = scorefxn(tmp_pose)

        pose_list.append(tmp_pose.clone())
        energy_list.append({
            "Index": i + 1,
            "Energy_Before_MC": energy_before_mc,
            "Energy_After_MC": energy_after_mc
        })

    # === Write CSV ===
    with open(out_csv_path, "w", newline="") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=["Index", "Energy_Before_MC", "Energy_After_MC"])
        writer.writeheader()
        writer.writerows(energy_list)

    # === Write Multi-model PDB ===
    with open(out_pdb_path, "w") as f:
        for i, p in enumerate(pose_list):
            f.write(f"MODEL     {i + 1}\n")
            f.write(pose_to_string(p))
            f.write("ENDMDL\n")

    print(f"Saved {len(pose_list)} total conformations (including original) to '{out_pdb_path}'")
    print(f"Energies written to '{out_csv_path}'")

# === Example Usage ===
if __name__ == "__main__":
    relaxPose = pose_from_pdb('/Users/alecloftus/Desktop/PKHLAB/PV_Reengagement/1rwy_WT/1rwy_wt.pdb')
    scorefxn = get_score_function()

    print("Original sequence:", relaxPose.sequence())
    print("Original energy:", scorefxn(relaxPose))

    pack_and_save_multiple_rotamers(
        pose=relaxPose,
        posi=45,
        amino='A',
        scorefxn=scorefxn,
        n_structs=30,
        out_pdb_path='1rwy_A45_diverse_rotamers.pdb',
        out_csv_path='1rwy_A45_energies.csv',
        mc_temp=2.0,
        apply_backbone_perturb=True
    )
```
